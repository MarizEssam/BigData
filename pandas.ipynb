# DataFrame, Pipes

```python
def removena(df):
    return df.dropna()
def fixdate(df):
    dfc = df.copy()
    dfc['Date'] = pd.to_datetime(dfc['Date'])
    return dfc
def removeduplicates(df):
    dfc = df.copy()
    dfc.drop_duplicates()
    return dfc
```

```python
df = pd.read_csv('data/wdata.csv')
df
```

```python
dfc = df.pipe(removena).pipe(fixdate).pipe(removeduplicates)
dfc
```

```python
def clean(df):
    dfc = df.copy()
    dfc = dfc.dropna()
    dfc['Date'] = pd.to_datetime(dfc['Date'])
    dfc.drop_duplicates()
    return dfc
def modify(df, value):
    dfc = df.copy()
    dfc['Final'] = value
    return dfc
```

```python
dfc = df.pipe(clean).pipe(modify, 55)
dfc
```
