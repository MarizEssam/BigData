# Print Timetable Example

```python
def print_times_tabe(x):
	print('-------------')
	for m in range(0,11):
		r = x * m;
		print("%d * %d \t= %d"%(x, m, r));
```

```python
print_times_tabe(7)
```

# Nested Functions

```python
def alt_print_times_tabe(x):
	def print_times(x, m):
		r = x * m;
		print("%d * %d \t= %d"%(x, m, r));
	print('-------------')
	for m in range(0,11):
		print_times(x, m)
```

```python
alt_print_times_tabe(8)
```

# Classes

```python
class BasicStudent(object):	
	def __init__(self, first_name, last_name):
		self.first_name = first_name; 
		self.last_name = last_name;
		print(self.__class__.__name__, "constructed")
	def __del__(self):
		print(self.__class__.__name__, "destroyed")
	def __str__(self):
		return self.name()
	def name(self):
		return self.first_name + ' ' + self.last_name;
```

```python
student = BasicStudent('ahmeD', 'salem');
```

```python
student.name()
```

```python
print(student)
```

# Inheritance

```python
class Student(BasicStudent):
	def __init__(self, first_name, last_name, age):
		super(Student, self).__init__(first_name, last_name)
		self.age = age
		print(self.__class__.__name__, "constructed")
	def name(self):
		return (self.first_name + ' ' + self.last_name).title();
	def info(self):
		return self.name() + ', '+ str(self.age) + ' year';
	def __del__(self):
		print(self.__class__.__name__, "destroyed")
```

```python
student = Student('ahmeD', 'salem', 18);
```

```python
student.name()
```

```python
student.info()
```

```python
print(student)
```

# Define Iterator: Primes

```python
class Primes:
	def __init__(self, max):
		self.count = max
	
	def __iter__(self):
		self.prime = 2
		return self
	def __is_prime(self, test):
		x = test - 1
		while(x>1):
			if((test%x)==0):
				return False
			x -= 1
		return True
	def __next__(self):
		x = self.prime
		self.prime += 1
		self.count -= 1
		if(self.count==0): raise StopIteration;
		while(not self.__is_prime(self.prime)):
			self.prime += 1
		return x
```

```python
primes = Primes(100)
```

```python
for x in primes:
	print(x)
```
